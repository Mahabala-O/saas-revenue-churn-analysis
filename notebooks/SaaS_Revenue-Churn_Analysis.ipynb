```python
## churn rate by plan ##
churn_by_plan = subscriptions[subscriptions['churned'] == 'Yes']
```

```python
churn_by_plan.head()
```

```python
data_columns = ['plan', 'industry', 'acquisition_channel', 'region', 'churn_reason', 'company_size']
```

```python
## calculating churns by different categories and reasons ## 
for col in data_columns: 
    print(f"\n{col}:")
    print(churn_by_plan[col].value_counts())
```

```python
for col in data_columns:
    display(churn_by_plan[col].value_counts().to_frame())
    print("-" * 30)
```

```python
for col in data_columns: 
    total = subscriptions.groupby(col).size().rename('total')
    churned = subscriptions[subscriptions['churned']=='Yes'].groupby(col).size().rename('churned_count')
    churn_rate = (churned/total*100).round(2).rename('churn_rate_%')
    
    display(pd.concat([total, churned, churn_rate], axis=1).sort_values('churn_rate_%', ascending=False))
    print("-" * 30)
```
