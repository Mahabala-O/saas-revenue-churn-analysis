```python
##  net revenue retention (new MRR minus churned MRR) ##
MRR_Trends['net_revenue_retention'] = MRR_Trends['total_mrr'] - (MRR_Trends['churned_customers'] * MRR_Trends['avg_revenue_per_customer'])
MRR_Trends['churned_loss'] = MRR_Trends['churned_customers'] * MRR_Trends['avg_revenue_per_customer']
```

```python
mean_nrr = MRR_Trends['net_revenue_retention'].mean()
std_nrr = MRR_Trends['net_revenue_retention'].std()

spikes = MRR_Trends[MRR_Trends['net_revenue_retention'] > mean_nrr + 1.5*std_nrr]
dips = MRR_Trends[MRR_Trends['net_revenue_retention'] < mean_nrr - 1.5*std_nrr]

cols = ['month_dt', 'total_mrr', 'net_revenue_retention', 'churned_loss', 'new_customers', 'churned_customers']

print("Spikes:")
display(spikes[cols])
print("Dips:")
display(dips[cols])
```

```python
phases = {
    'Growth': MRR_Trends[MRR_Trends['month_dt'] < '2023-07-01'],
    'Scaling': MRR_Trends[(MRR_Trends['month_dt'] >= '2023-07-01') & (MRR_Trends['month_dt'] < '2024-07-01')],
    'Mature': MRR_Trends[MRR_Trends['month_dt'] >= '2024-07-01']
}

cols = ['month_dt', 'total_mrr', 'net_revenue_retention', 'churned_loss', 'new_customers', 'churned_customers']

for phase_name, phase_df in phases.items():
    mean_nrr = phase_df['net_revenue_retention'].mean()
    std_nrr = phase_df['net_revenue_retention'].std()
    
    spikes = phase_df[phase_df['net_revenue_retention'] > mean_nrr + 1.5*std_nrr]
    dips = phase_df[phase_df['net_revenue_retention'] < mean_nrr - 1.5*std_nrr]
    
    print(f"\n=== {phase_name} Phase ===")
    print("Spikes:"), display(spikes[cols])
    print("Dips:"), display(dips[cols])
```

```python
today = pd.Timestamp.today()

subscriptions['tenure_days'] = np.where(
    subscriptions['churned'] == 'Yes',
    (subscriptions['churn_date_dt'] - subscriptions['signup_date_dt']).dt.days,
    (today - subscriptions['signup_date_dt']).dt.days
)

subscriptions['tenure_months'] = subscriptions['tenure_days'] / 30.44
```

```python
# Normalize each variable to 0-1 scale (higher = more at risk)
subscriptions['risk_feature'] = 1 - (subscriptions['feature_usage_pct'] / 100)
subscriptions['risk_nps'] = 1 - (subscriptions['nps_score'] / 10)
subscriptions['risk_tenure'] = 1 - (subscriptions['tenure_months'] / subscriptions['tenure_months'].max())

# Weighted composite score
subscriptions['risk_score'] = (
    subscriptions['risk_feature'] * 0.5 +
    subscriptions['risk_nps'] * 0.3 +
    subscriptions['risk_tenure'] * 0.2
)
```

```python
churned_risk_mean = subscriptions[subscriptions['churned'] == 'Yes']['risk_score'].mean()

at_risk_composite = subscriptions[
    (subscriptions['churned'] == 'No') &
    (subscriptions['risk_score'] > churned_risk_mean)
]

print(f"Composite threshold: {churned_risk_mean:.3f}")
print(f"At-risk customers: {len(at_risk_composite)}")
print(f"As % of active base: {len(at_risk_composite) / len(subscriptions[subscriptions['churned'] == 'No']) * 100:.1f}%")
```

```python
# See how churned vs active customers score on the composite
subscriptions.groupby('churned')['risk_score'].mean()
```

```python
import matplotlib.pyplot as plt

subscriptions[subscriptions['churned'] == 'Yes']['risk_score'].hist(alpha=0.5, label='Churned', bins=20)
subscriptions[subscriptions['churned'] == 'No']['risk_score'].hist(alpha=0.5, label='Active', bins=20)
plt.legend()
plt.title('Risk Score Distribution')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

y_true = (subscriptions['churned'] == 'Yes').astype(int)
auc = roc_auc_score(y_true, subscriptions['risk_score'])
print(f"AUC Score: {auc:.3f}")
```
